```python
data['BMI_Category']=data['BMI_Category'].str.replace('Normal Weight','Normal')
```

```python
data.head(20)
```

```python
data['BMI_Category'].value_counts().reset_index()
```

```python
data[['BP_1(systolic)','BP_2(diastolic)']] = data.Blood_Pressure.str.split("/",expand=True)
data
data['BP_1(systolic)'] = pd.to_numeric(data['BP_1(systolic)'])
data['BP_2(diastolic)'] = pd.to_numeric(data['BP_2(diastolic)'])
data.sample(5)
```

```python
blood_class_conditions = [
    (data['BP_1(systolic)'] < 120) & (data['BP_2(diastolic)'] < 80),
    (data['BP_1(systolic)'].between(120, 140)) & (data['BP_2(diastolic)'] < 90),
    (data['BP_1(systolic)'] >= 140) & (data['BP_2(diastolic)'] >= 90) | (data['BP_2(diastolic)'] >= 80)
]

labels = ['Optimal', 'Normal', 'Hypertension']

data['Blood Pressure Category'] = np.select(blood_class_conditions, labels, default='Undefined')
```

```python
data
```

```python
#From the results, we have some considerations:

# Saleperson could be merged with Sales Representative

# Software Engineer could be merged with Engineer
```

```python
data["Occupation"] = data["Occupation"].replace("Sales Representative", "Salesperson")

data["Occupation"] = data["Occupation"].replace("Software Engineer","Engineer")
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
sns.boxplot(x = "Gender",y = "Age",data = data,hue = 'Gender',ax = axs[0]).set(title = 'How old they are')
sns.swarmplot(x="Gender", y="Sleep_Duration", data=data,ax=axs[1]).set_title("2. How long do they sleep?")
sns.barplot(x="Gender", y="Quality_of_Sleep", data=data,ax=axs[2]).set_title("3. How well do they sleep?")
plt.show()
```

```python
ax = sns.boxplot(x = 'BMI_Category',y = 'Age',data = data,hue = 'BMI_Category').set_title ('BMI_cat VS Age')
```

```python
correlation_table = data[["Quality_of_Sleep", "Sleep_Duration", "Age", "Physical_Activity_Level", "Stress_Level", "Heart_Rate", "Daily_Steps"]].corr()

plt.figure(figsize=(16, 5))
heatmap = sns.heatmap(correlation_table, annot=True, cmap='Blues')
heatmap.set_title('Correlation Heatmap');
```
